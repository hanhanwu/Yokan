```python
from datasets import load_dataset
from pprint import pprint
```

### Wix/WixQA 🌟

```python
qa_ds = load_dataset("Wix/WixQA", "wixqa_expertwritten")
qa_corpous = load_dataset("Wix/WixQA", "wix_kb_corpus")
print(qa_ds.shape, qa_corpous.shape)
```

```python
# Inspect dataset structure and sample records
print(qa_ds)
print("Splits:", list(qa_ds.keys()))
for split in qa_ds:
    ds = qa_ds[split]
    print(f"\n=== Split: {split} ===")
    print("Rows:", len(ds))
    print("Columns:", ds.column_names)
    print("Features:", ds.features)
    print("First record:")
    pprint(ds[0])
    break

print(qa_corpous)
print("Splits:", list(qa_corpous.keys()))
for split in qa_corpous:
    ds = qa_corpous[split]
    print(f"\n=== Split: {split} ===")
    print("Rows:", len(ds))
    print("Columns:", ds.column_names)
    print("Features:", ds.features)
    print("First record:")
    pprint(ds[0])
    break
```

```python
import os

output_dir = os.path.join("wix_docs")
os.makedirs(output_dir, exist_ok=True)

for record in qa_corpous['train']:
    file_path = os.path.join(output_dir, f"{record['id']}.txt")
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(record["contents"])

print(f"Saved {len(qa_corpous['train'])} documents to {output_dir}")
```
